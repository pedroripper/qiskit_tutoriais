```python
from qiskit import*
%matplotlib notebook
# matplotlib notebook pode rotacionar a esfera
```

```python
qc = QuantumCircuit(2,2)
qc.h(0)
qc.draw(output='mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
job = execute(qc, backend)
result = job.result()
statevec =  result.get_statevector()
```

```python
from qiskit.tools.visualization import plot_bloch_multivector

plot_bloch_multivector(statevec)
```

```python
# |phi+> = (|00> + |11>)/sqrt(2)

# Controlled not
# Target Control

qc.cx(0,1)
qc.draw(output='mpl')
```

```python
# T C
# 0 0 -> 00
# 0 1 -> 11
# 1 0 -> 10
# 1 1 -> 01
```

```python
import numpy as np

qc2 = QuantumCircuit(1)

qc2.rx(np.pi/2,0)
qc2.z(0)
qc2.draw(output='mpl')
```

```python
qc4 = QuantumCircuit(1)
qc4.z(0)
qc4.rx(-np.pi/2,0)
qc4.draw(output='mpl')
```

```python
qc.z(0)
qc.z(1)
qc.rx(-np.pi/2,0)

qc.rx(-np.pi/2,1)

qc.draw(output='mpl')
```

```python
qc.measure([0,1],[0,1])
qc.draw(output='mpl')
```
